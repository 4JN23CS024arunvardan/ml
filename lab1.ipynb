```python
numerical_feature = data.select_dtypes (include =[np.number]).columns
print(numerical_feature)
```

```python
plt.figure(figsize=(15,10))
for i,feature in enumerate(numerical_feature):
    plt.subplot(3,3,i+1)
    sns.histplot(data[feature],kde =True,bins=50,color='green')
    plt.title(f'distribution of {feature}')
plt.tight_layout()
```

```python
plt.figure(figsize=(15,10))
for i,feature in enumerate(numerical_feature):
    plt.subplot(3,3,i+1)
    sns.boxplot(data[feature],color='pink')
    plt.title(f'distribution of {feature}')
plt.tight_layout()
```

```python
print("Outliner Detection:")
outliers_summary={}
for feature in numerical_feature:
    Q1=data[feature].quantile(0.25)
    Q3=data[feature].quantile(0.75)
    IQR =Q3 - Q1
    lower_bound =Q1 - 1.5*IQR
    upper_bound =Q1 + 1.5*IQR
    outliers=data[(data[feature]<lower_bound) | (data[feature] > upper_bound)]
    outliers_summary[feature]=len(outliers)
    print(f"{feature}:{len(outliers)}outliers")
```

```python
print("\nDataset Summary:")
print(data.describe().T)
```

```python
correlation_matrix =data[numerical_feature].corr()
print(correlation_matrix)
```

```python
plt.figure (figsize=(15 , 10))
sns.heatmap(correlation_matrix, annot=True,cmap='coolwarm',
            fmt='.2f',linewidths=0.2)
plt.title('Correlation Matrix of California Housing Features')
plt.show()
```

```python
sns.pairplot(data, diag_kind='kde',plot_kws={'alpha':0.5})
plt.suptitle('Pair Plot of California Hosuing features',y=1.02)
plt.show()
```
